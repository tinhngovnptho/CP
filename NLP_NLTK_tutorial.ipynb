# **Stemming**
Stemming là quá trình loại bỏ một phần của từ, hoặc tối giản một từ thành gốc của nó
[document](https://www.nltk.org/howto/stem.html)

**Porter Stemmer**

```python
from nltk.stem import PorterStemmer

words = ["play", "playing", "plays", "played"]

for words in words:
  print(words + ": " + PorterStemmer().stem(words))
```

```python
words = ["running", "runner", "flies", "flying", "easily", "happier"]

for words in words:
  print(words + ": " + PorterStemmer().stem(words))
```

**Lancaster Stemmer**

```python
from nltk.stem import LancasterStemmer

words = ["running", "runner", "flies", "flying", "easily", "happier"]

for words in words:
  print(words + ": " + LancasterStemmer().stem(words))
```

**Snowball Stemmer**

```python
from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer("english")

words = ["running", "runner", "flies", "flying", "easily", "happier"]

for words in words:
  print(words + ": " + stemmer.stem(words))
```

**Over stemming**

Là quá trình mà một phần lớn hơn của từ bị cắt bỏ nhiều hơn so với yêu cầu

```python
words = ["universe", "university"]

for words in words:
  print(words + ": " + PorterStemmer().stem(words))
```

**Under stemming**

Under stemming xảy ra khi stemmer không nhóm các từ có cùng gốc lại với nhau

```python
words = ["ran", "running", "run"]

for words in words:
  print(words + ": " + PorterStemmer().stem(words))
```

# **Lemmatization (Biến thể gốc từ)**
Lemmatization rút gọn từ về dạng cơ sở (lemma), và luôn trả về từ có nghĩa
[document](https://www.nltk.org/api/nltk.stem.wordnet.html)

```python
from nltk.stem import WordNetLemmatizer

# Khởi tạo bộ Lemmatization
lemmatizer = WordNetLemmatizer()

# Danh sách từ để thử nghiệm
words = ["running", "ran", "flies", "better", "easily", "happier"]

print("Lemmatization:")
for word in words:
    print(f"{word} -> {lemmatizer.lemmatize(word, pos='v')}")  # 'v' cho động từ
print('-------')
for word in words:
   print(f"{word} -> {lemmatizer.lemmatize(word, pos='a')}")
```

# **Stop Words**
Lọc các từ không mang nhiều ý nghĩa
[document](https://www.geeksforgeeks.org/removing-stop-words-nltk-python/)

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words("english"))
print("Danh sách các stop word:")
print(stop_words)

text = "This is a simple example to demonstrate stop words removal in text preprocessing."

words = word_tokenize(text)

filtered_words = []  # Khởi tạo một danh sách rỗng để lưu các từ đã lọc

# Explicit loop
for word in words:
    if word.lower() not in stop_words:
        filtered_words.append(word)

# List comprehension
# filtered_words = [word for word in words if word.lower() not in stop_words]

print("\nText after removing stop words:")
print(" ".join(filtered_words))
```

# **Part of Speech (POS) tagging**
POS Tagging (Part of Speech Tagging) hay gán nhãn từ loại là quá trình gán nhãn từ loại
documents: [document](https://www.nltk.org/api/nltk.tag.pos_tag.html)

image.png

```python
from nltk import pos_tag

# Văn bản mẫu
text = "The quick brown fox jumps over the lazy dog."

# Tokenization
words = word_tokenize(text)

print("Từ đã được tách:")
print(words)

# Gán nhãn từ loại
pos_tags = pos_tag(words)
print("\nKết quả POS Tagging:")
print(pos_tags)
```

image.png

```python
texts = [
    "The back door",
    "On my back",
    "Win the voters back",
    "Promised to back the bill"
]

for text in texts:
    print(f"{text}: {pos_tag(word_tokenize(text))}")
```

# **Chunking**
Chunking (phân đoạn hoặc phân cụm) là quá trình nhóm các từ trong một câu thành cụm từ ngữ pháp có ý nghĩa
[document](https://www.nltk.org/api/nltk.chunk.html)

```python
import nltk
from nltk import word_tokenize, pos_tag, RegexpParser

# Văn bản mẫu
text = "The quick brown fox jumps over the lazy dog."

# Tokenization và POS Tagging
words = word_tokenize(text)
pos_tags = pos_tag(words)  # Gắn nhãn từ loại
print("POS Tagging:")
print(pos_tags)

# Định nghĩa cú pháp chunking (cụm danh từ NP)
grammar = """
  NP: {<DT>?<JJ.*>*<NN.*>+}   # Mạo từ + Tính từ (tùy chọn) + Danh từ (NN có thể là số ít, số nhiều)
"""

# Chunking
chunk_parser = RegexpParser(grammar)
tree = chunk_parser.parse(pos_tags)

# Hiển thị kết quả
print("\nKết quả Chunking:")
print(tree)
```

# **NER (Named Entity Recognition)**
Là một tác vụ nhằm xác định và phân loại các thực thể được nhắc đến trong văn bản vào các danh mục cụ thể
[document](https://www.nltk.org/book/ch07.html)


image.png

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import ne_chunk

# Văn bản mẫu
text = "Donald Trump was born in New York and was the 47th President of the United States."

# Tokenization và POS Tagging
words = word_tokenize(text)
pos_tags = pos_tag(words)

print("POS Tagging:")
print(pos_tags)

# NER: Nhận diện thực thể có tên
named_entities = ne_chunk(pos_tags, binary=False)  # binary=False để phân loại chi tiết
print("\nKết quả NER:")
print(named_entities)
```

# **Sentiment analysis**

```python
from nltk.sentiment import SentimentIntensityAnalyzer


# Tải VADER (bộ từ điển cảm xúc được thiết kế đặc biệt để phân tích cảm xúc trong văn bản tiếng Anh)
nltk.download('vader_lexicon')

# Khởi tạo bộ phân tích cảm xúc
sia = SentimentIntensityAnalyzer()

# Một số câu ví dụ
texts = [
    "I love this product! It's amazing",
    "This is the worst movie ever. Terrible!",
    "It's okay, not the best but not the worst.",
    "This movie is so bad it's actually good"
]

# Chạy phân tích và in kết quả
for text in texts:
    score = sia.polarity_scores(text)
    print(f"Text: {text}")
    print(f"Sentiment Scores: {score}")
    print("-" * 40)
```
